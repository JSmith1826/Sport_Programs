```python
# Import Dependencies

import pandas as pd

```

```python
# read the first file created into a dataframe

### The file is the Baseball Yearbooks grid that runs from 1950 to 1979. I manually entered the data 7/12/22
bb_df = pd.read_csv('data/BB-YB-01_manual.csv')
bb_df.head()
```

```python
# Create a list from 1950 to 1979 to hand as column names

def createList(n):
    column_list = []
    for i in range(n+1):
        column_list.append(1950+i)
    return(column_list)

# print(createList(29))

column_names = str(createList(29))
column_names
```

```python
# Tranform string column


def change_string(member):
    if member == 'x':
        return 1
    else:
        return 0
    
 
    
bb_df[column_names] = bb_df[column_names].apply(change_string)
bb_df.head()
```
